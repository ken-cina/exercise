```python
# 1 Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2 Load data from SPX file to df
# 2.1 Use Date in the index
df_SPX = pd.read_csv('https://s3.eu-west-1.amazonaws.com/neueda.conygre.com/pydata/SPX.csv',
                     index_col='Date', parse_dates=True)

# df_SPX.info()
df_SPX.head()
```

```python
# 3 Convert all columns to numerical
# 3.1 Clean the columns with .str.replace()
# df_SPX['Date'].str.replace('','')
df_SPX['Price'] = df_SPX['Price'].str.replace(',','')
df_SPX['Open'] = df_SPX['Open'].str.replace(',','')
df_SPX['High'] = df_SPX['High'].str.replace(',','')
df_SPX['Low'] = df_SPX['Low'].str.replace(',','')
df_SPX['Change %'] = df_SPX['Change %'].str.replace('%','')

# 3.2 Convert to numerrical with pd.to_numerical() 
df_SPX['Price'] = pd.to_numeric(df_SPX['Price'])
df_SPX['Open'] = pd.to_numeric(df_SPX['Open'])
df_SPX['High'] = pd.to_numeric(df_SPX['High'])
df_SPX['Low'] = pd.to_numeric(df_SPX['Low'])
df_SPX['Change %'] = pd.to_numeric(df_SPX['Change %'])

df_SPX
```

```python
# 4 Sort the dataframe in chronological order
df_SPX = df_SPX.sort_values('Date')

df_SPX
```

```python
# 5 Plot a line chart: June 2015 to June 2016, High + Low
cols = ['High', 'Low']
df_SPX['June 2015':'June 2016'][cols].plot(kind='line', figsize=(10,6))
```
