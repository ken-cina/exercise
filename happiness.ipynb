```python
# import libries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data from data files
work_dir = "./"
# files_list = ["2015","2016","2017","2018","2019","2020","2021","2022"]
df_2015 = pd.read_csv(work_dir+"2015"+".csv")
df_2016 = pd.read_csv(work_dir+"2016"+".csv")
df_2017 = pd.read_csv(work_dir+"2017"+".csv")
df_2018 = pd.read_csv(work_dir+"2018"+".csv")
df_2019 = pd.read_csv(work_dir+"2019"+".csv")
df_2020 = pd.read_csv(work_dir+"2020"+".csv")
df_2021 = pd.read_csv(work_dir+"2021"+".csv")
df_2022 = pd.read_csv(work_dir+"2022"+".csv")

# show table
df_2022
# df.info()
```

```python
# data processing

## select top 10 highest happiness score country from dataframe

countries_list_2015 = df_2015['Country'][0:10].to_list()
scores_list_2015 = df_2015['Happiness Score'][0:10].to_list()
zipped_2015 = list(zip(countries_list_2015, scores_list_2015))

countries_list_2016 = df_2016['Country'][0:10].to_list()
scores_list_2016 = df_2016['Happiness Score'][0:10].to_list()
zipped_2016 = list(zip(countries_list_2016, scores_list_2016))

countries_list_2017 = df_2017['Country'][0:10].to_list()
scores_list_2017 = df_2017['Happiness.Score'][0:10].to_list()
zipped_2017 = list(zip(countries_list_2017, scores_list_2017))

countries_list_2018 = df_2018['Country or region'][0:10].to_list()
scores_list_2018 = df_2018['Score'][0:10].to_list()
zipped_2018 = list(zip(countries_list_2018, scores_list_2018))

# countries_list_2019 = df_2019['Country'][0:10].to_list()
# scores_list_2019 = df_2019['Happiness score'][0:10].to_list()
# zipped_2019 = list(zip(countries_list_2019, scores_list_2019))

countries_list_2019 = df_2019['Country or region'][0:10].to_list()
scores_list_2019 = df_2019['Score'][0:10].to_list()
zipped_2019 = list(zip(countries_list_2019, scores_list_2019))

countries_list_2020 = df_2020['Country name'][0:10].to_list()
scores_list_2020 = df_2020['Ladder score'][0:10].to_list()
zipped_2020 = list(zip(countries_list_2020, scores_list_2020))

countries_list_2021 = df_2021['Country name'][0:10].to_list()
scores_list_2021 = df_2021['Ladder score'][0:10].to_list()
zipped_2021 = list(zip(countries_list_2021, scores_list_2021))

countries_list_2022 = df_2022['Country'][0:10].to_list()
scores_list_2022 = pd.to_numeric(df_2022['Happiness score'][0:10].str.replace(',','')).to_list()
zipped_2022 = list(zip(countries_list_2022, scores_list_2022))


country_scores = {}
for item in zipped_2015:
    if item[0] not in country_scores.keys():
        country_scores[item[0]] = item[1]
    else:
        country_scores[item[0]] += item[1]
for item in zipped_2016:
    if item[0] not in country_scores.keys():
        country_scores[item[0]] = item[1]
    else:
        country_scores[item[0]] += item[1]
for item in zipped_2017:
    if item[0] not in country_scores.keys():
        country_scores[item[0]] = item[1]
    else:
        country_scores[item[0]] += item[1]
for item in zipped_2018:
    if item[0] not in country_scores.keys():
        country_scores[item[0]] = item[1]
    else:
        country_scores[item[0]] += item[1]
for item in zipped_2019:
    if item[0] not in country_scores.keys():
        country_scores[item[0]] = item[1]
    else:
        country_scores[item[0]] += item[1]
        
# 2020
for item in zipped_2020:
    if item[0] not in country_scores.keys():
        country_scores[item[0]] = item[1]
    else:
        country_scores[item[0]] += item[1]
for item in zipped_2021:
    if item[0] not in country_scores.keys():
        country_scores[item[0]] = item[1]
    else:
        country_scores[item[0]] += item[1]

# 2022
for item in zipped_2022:
    if item[0] not in country_scores.keys():
        country_scores[item[0]] = item[1]/1000
    else:
        country_scores[item[0]] += item[1]/1000
        
# show
country_scores
sorted(country_scores.items(), key = lambda x:x[1], reverse=True)
```

```python
# import wordcloud to plot the wordcloud chart
from wordcloud import WordCloud

# cloud_text = ' '.join(country_list)
cloud_text = country_scores

# word cloud generated from text
# word_cloud = WordCloud().generate(newtxt)

# word cloud generated from frequencies(dictionary)
word_cloud = WordCloud(max_words=100,
        width=2000,
        height=1200,
#         contour_width = 4,
#         contour_color = 'steelblue',
        background_color='white'
        max_font_size=50,
        min_font_size=10,
#         mode='RGBA',
        #colormap='pink'
                  ).generate_from_frequencies(cloud_text)

word_cloud.to_file('word_could_chart.jpg')

# show word cloud chart file
plt.imshow(word_cloud)
plt.axis("off")
plt.show()
```
